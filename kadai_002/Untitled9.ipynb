```python
from bs4 import BeautifulSoup
from urllib import request
import re
import MeCab

#mecabの初期化
mecab = MeCab.Tagger('-Owakati')

#ストップワードのセットを定義
stop_words = set(["の", "は", "ため", "です", "ます", "する", "れる", "こと", "これ", "それ"])

#スクレイピングの準備
url = 'https://www.aozora.gr.jp/cards/000148/files/2371_13943.html'
response = request.urlopen(url)
soup = BeautifulSoup(response)
response.close()

#タグの削除
natume_text = soup.find('div', class_='main_text')
natume_text = natume_text.text

#改行コードなど削除
natume_text = re.sub(r'[\u3000 \n \r]', '', natume_text)
natume_text

#単語に分割
words = mecab.parse(natume_text).split()

#ストップワードの除去
stop_text = [word for word in words if word not in stop_words]

#リストを文字列にする
filtered_text = ' '.join(stop_text)
print(filtered_text)
```
